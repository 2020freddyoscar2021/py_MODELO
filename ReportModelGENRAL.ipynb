```python
#Se calcula los Pasajeros por Kilometros
flw24['Pasxkm'] = (flw24['To_MP'] - flw24['From_MP'])*flw24['TransitFlow']
flw25['Pasxkm'] = (flw25['To_MP'] - flw25['From_MP'])*flw25['TransitFlow']
#flw30['Pasxkm'] = (flw30['To_MP'] - flw30['From_MP'])*flw30['TransitFlow']
#flw35['Pasxkm'] = (flw35['To_MP'] - flw35['From_MP'])*flw35['TransitFlow']
#flw38['Pasxkm'] = (flw38['To_MP'] - flw38['From_MP'])*flw38['TransitFlow']
#flw40['Pasxkm'] = (flw40['To_MP'] - flw40['From_MP'])*flw40['TransitFlow']
#flw24.head()
```

# Solo se remplaza el nombre Route_ID por ROUTE, para la union de tablas

```python
df_Routes.rename(columns={'Route_ID': 'ROUTE'}, inplace=True)
df_Routes.head()
```

```python
TRutas = df_Routes[['ROUTE','NombreCorto','NombreLargo','ModoRutaHPM','FrecHPM','capHPM']]
TRutas.head()
```

```python
ono_sum24 = ono24.groupby('ROUTE').sum()[['On']]
ono_sum25 = ono25.groupby('ROUTE').sum()[['On']]
#ono_sum30 = ono30.groupby('ROUTE').sum()[['On']]
#ono_sum35 = ono35.groupby('ROUTE').sum()[['On']]
#ono_sum38 = ono38.groupby('ROUTE').sum()[['On']]
#ono_sum40 = ono40.groupby('ROUTE').sum()[['On']]
ono_sum24.head()
```

```python
flw_max24 = flw24.groupby('Route').max()[['TransitFlow']]
flw_max25 = flw25.groupby('Route').max()[['TransitFlow']]
#flw_max30 = flw30.groupby('Route').max()[['TransitFlow']]
#flw_max35 = flw35.groupby('Route').max()[['TransitFlow']]
#flw_max38 = flw38.groupby('Route').max()[['TransitFlow']]
#flw_max40 = flw40.groupby('Route').max()[['TransitFlow']]
flw_max24.head()
```

```python
flw_Paskm24 = flw24.groupby('Route').sum()[['Pasxkm']]
flw_Paskm25 = flw25.groupby('Route').sum()[['Pasxkm']]
#flw_Paskm30 = flw30.groupby('Route').sum()[['Pasxkm']]
#flw_Paskm35 = flw35.groupby('Route').sum()[['Pasxkm']]
#flw_Paskm38 = flw38.groupby('Route').sum()[['Pasxkm']]
#flw_Paskm40 = flw40.groupby('Route').sum()[['Pasxkm']]
flw_Paskm24.head()
```

```python
flw_sumIVTT24 = flw24.groupby('Route').sum()[['BaseIVTT']]
flw_sumIVTT25 = flw25.groupby('Route').sum()[['BaseIVTT']]
#flw_sumIVTT30 = flw30.groupby('Route').sum()[['BaseIVTT']]
#flw_sumIVTT35 = flw35.groupby('Route').sum()[['BaseIVTT']]
#flw_sumIVTT38 = flw38.groupby('Route').sum()[['BaseIVTT']]
#flw_sumIVTT40 = flw40.groupby('Route').sum()[['BaseIVTT']]
flw_sumIVTT24.head()
```

```python
flw_km24 = flw24.groupby('Route').max()[['To_MP']]
flw_km25 = flw25.groupby('Route').max()[['To_MP']]
#flw_km30 = flw30.groupby('Route').max()[['To_MP']]
#flw_km35 = flw35.groupby('Route').max()[['To_MP']]
#flw_km38 = flw38.groupby('Route').max()[['To_MP']]
#flw_km40 = flw40.groupby('Route').max()[['To_MP']]
flw_km24.head()
```

```python
df_mergeonoflw24 = pd.merge(ono_sum24, flw_max24, how='left', left_index = True, right_index = True)
df_mergeonoflwIVTT24 = pd.merge(df_mergeonoflw24, flw_sumIVTT24, how='left', left_index = True, right_index = True)
df_mergeGeneral24 = pd.merge(df_mergeonoflwIVTT24, flw_km24, how='left', left_index = True, right_index = True)
df_mergeGeneralB24 = pd.merge(df_mergeGeneral24, flw_Paskm24, how='left', left_index = True, right_index = True)
df_mergeonoR24 = pd.merge(df_mergeGeneralB24,TRutas, on = ['ROUTE'], how='inner')
```

```python
df_mergeonoflw25 = pd.merge(ono_sum25, flw_max25, how='left', left_index = True, right_index = True)
df_mergeonoflwIVTT25 = pd.merge(df_mergeonoflw25, flw_sumIVTT25, how='left', left_index = True, right_index = True)
df_mergeGeneral25 = pd.merge(df_mergeonoflwIVTT25, flw_km25, how='left', left_index = True, right_index = True)
df_mergeGeneralB25 = pd.merge(df_mergeGeneral25, flw_Paskm25, how='left', left_index = True, right_index = True)
df_mergeonoR25 = pd.merge(df_mergeGeneralB25,TRutas, on = ['ROUTE'], how='inner')
```
