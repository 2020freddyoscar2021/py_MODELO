# Se Calcula Los Pasajeros por Kilometro

```python
flw['Pasxkm'] = (flw['To_MP'] - flw['From_MP'])*flw['TransitFlow']
flw.head()
```

*Solo se remplaza el nombre Route_ID por ROUTE, para la union de tablas

```python
Routes.rename(columns={'Route_ID': 'ROUTE'}, inplace=True)
Routes.head()
```

# Se extrae solo los campos necesarios

```python
TRutas = Routes[['ROUTE','NombreCorto','NombreLargo','ModoRutaHPM','FrecHPM','capHPM']]
TRutas.head()
```

# Se agrupan campos, se crean tablas resumidas

```python
ono_sum = ono.groupby('ROUTE').sum()[['On']]
ono_sum.head()
```

```python
flw_max = flw.groupby('Route').max()[['TransitFlow']]
flw_max.head()
```

```python
flw_Paskm = flw.groupby('Route').sum()[['Pasxkm']]
flw_Paskm.head()
```

```python
flw_sumIVTT = flw.groupby('Route').sum()[['BaseIVTT']]
flw_sumIVTT.head()
```

```python
flw_km = flw.groupby('Route').max()[['To_MP']]
flw_km.head()
```

# Se realiza el merge , union de tablas

```python
df_mergeonoflw = pd.merge(ono_sum, flw_max, how='left', left_index = True, right_index = True)
df_mergeonoflwIVTT = pd.merge(df_mergeonoflw, flw_sumIVTT, how='left', left_index = True, right_index = True)
df_mergeGeneral = pd.merge(df_mergeonoflwIVTT, flw_km, how='left', left_index = True, right_index = True)
df_mergeGeneral2 = pd.merge(df_mergeGeneral, flw_Paskm, how='left', left_index = True, right_index = True)
df_mergeonoR = pd.merge(df_mergeGeneral2,TRutas, on = ['ROUTE'], how='inner')
```

# Se realiza opearciones para calculo de indicadores de transporte

```python
df_mergeonoR['IndiceRotacion'] = df_mergeonoR['On']/df_mergeonoR['TransitFlow']
df_mergeonoR['Velocidad(km/h)'] = (df_mergeonoR['To_MP']/df_mergeonoR['BaseIVTT'])*60
df_mergeonoR['Vehxkm'] = (df_mergeonoR['To_MP']*60/df_mergeonoR['FrecHPM'])
df_mergeonoR['Frecuencia(veh/hora)'] = (60/df_mergeonoR['FrecHPM'])
df_mergeonoR['IPK'] = (df_mergeonoR['On']/df_mergeonoR['Vehxkm'])
df_mergeonoR['Flota Necesaria'] = (df_mergeonoR['BaseIVTT']*1.1*2/df_mergeonoR['FrecHPM'])
df_mergeonoR['Indice de Ocupacion'] = (df_mergeonoR['TransitFlow']/df_mergeonoR['capHPM'])
df_mergeonoR.head()
```

```python
df_mergeonoRSUM = df_mergeonoR.groupby('ModoRutaHPM').sum()[['On']]
df_mergeonoRSUM
```

```python
df_modoHPM = ModosHPM.merge(df_mergeonoRSUM, left_on='MODE_ID', right_on = 'ModoRutaHPM')
df_modoHPM.head()
```

# Se exporta formado Excel, se une pesta√±as

```python
with pd.ExcelWriter('py_ReportModel-'+anio+'-'+esc+'-'+per+'.xlsx') as writer:
    ono.to_excel(writer, index=True, sheet_name='ono')
    df_mergeonoR.to_excel(writer, index= True, sheet_name='General')
    df_modoHPM.to_excel(writer, index= True, sheet_name='Modo'+per)
    AnexoG.to_excel(writer, index= True, sheet_name='Compemento')
```
